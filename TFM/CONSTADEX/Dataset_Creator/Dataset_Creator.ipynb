## Creation of Datasets from Dicts using random normal distribution.
Note that new observations are batch-generated in controlled packs of typical case studies.
Each centre of the distribution and each standard deviation are controlled and 
assigned intentionally by a construction manager (the one who is scripting). 
We trust this expert data to furher enrich the DataBase rather than adding innecessary noise.
###  *Taking advantage of expert patterns avoiding multicollinearity.
With this method we try to induce some kind of "smart correlations" avoiding multicollinearity. Given the fact that the normal distribution functions are executed by variable within multiple different typical cases, the resulting dataset should be full of "expert rules" with little multicollinearity.

### Detached housing and other types:

```python
# Detached housing:

# Big area -> Big delay (Big problems tend to have lots of problems and interdepencies).
cons11_aisl=pd.DataFrame.from_dict({'modul_price':np.random.normal(910,375,200), 
                        "weeks_delay":np.random.normal(8,22,200),
                        "weeks_duration":np.random.normal(40,20,200),
                        "built_area": np.random.normal(200,150,200),
                        "typology":["DETACHED"]*200})

# Little area -> little delay (Little problems tend to be simpler and easier).
cons00_aisl=pd.DataFrame.from_dict({'modul_price':np.random.normal(720,200,800), 
                        "weeks_delay":np.random.normal(0,13,800),
                        "weeks_duration":np.random.normal(39,10,800),
                        "built_area": np.random.normal(100,200,800),
                        "typology":["DETACHED"]*800})

# Little module -> Big delay. (Budget assigned to the project too cheap to be true) .
cons001_aisl=pd.DataFrame.from_dict({'modul_price':np.random.normal(600,200,200), 
                        "weeks_delay":np.random.normal(8,22,200),
                        "weeks_duration":np.random.normal(33,15,200),
                        "built_area": np.random.normal(125,350,200),
                        "typology":["DETACHED"]*200})

const_tot_aisl = pd.concat([cons00_aisl,cons11_aisl,cons001_aisl])
```

```python
#Concat all the construction types
const_tot = pd.concat([ const_tot_aisl,const_tot_oth,const_tot_collec,const_tot_commerc])

#Let's sweep away negative values and other absurd values.
const_tot = const_tot[(const_tot["weeks_duration"]>13) & (const_tot["built_area"]>50)                       
                      & (const_tot['modul_price']>181) & (const_tot["weeks_delay"]>=0)]

#Weeks to integer type:
const_tot.weeks_duration= const_tot.weeks_duration.astype(int)
const_tot.weeks_delay= const_tot.weeks_delay.astype(int)

#reset index:
const_tot=const_tot.reset_index().drop("index",axis=1)
```

## One Hot Encoding (OHE) building typologies for better future processing:

```python
detached=[1 if n =="DETACHED" else 0 for n in const_tot["typology"]]
collective=[1 if n =="COLLECTIVE" else 0 for n in const_tot["typology"]]
commercial=[1 if n =="COMMERCIAL" else 0 for n in const_tot["typology"]]
others=[1 if n =="OTHERS" else 0 for n in const_tot["typology"]]

const_OHE = pd.DataFrame()
const_OHE["DETACHED"]  = pd.Series(detached)
const_OHE["COLLECTIVE"] = pd.Series(collective)
const_OHE["COMMERCIAL"] = pd.Series(commercial)
const_OHE["OTHERS"]= pd.Series(others)
```

#### One Hot Encoding (OHE) threshold:
The concept "True delay" depends on the delays and the duration, assigning a threshold based in the minimum proportion of 0.15 times the duration to consider it a TRUE delay:

```python
def divide(x):
    return np.divide(x['weeks_delay'],x['weeks_duration'])

delays = pd.DataFrame()
delays = const_tot.apply(divide,axis=1)
const_OHE["DELAYED"] = [1 if n > 0.2  else 0 for n in delays]
const_OHE.sample(5)
```
