```python
#Renombramos columnas:

colnm = list(softw)
def space_filler(namelist):
    nwnamelist=[]
    for i in namelist:
        nwname=''
        for j in i:
            if j==' ':
                nwname = nwname + '_'
            elif j=='.':
                nwname = nwname + ''
            elif j=='Á':
                nwname = nwname + 'A'
            elif j=='É':
                nwname = nwname + 'E'
            elif j=='Í':
                nwname = nwname + 'I'
            elif j=='Ó':
                nwname = nwname + 'O'
            elif j=='Ú':
                nwname = nwname + 'U'
            else:
                nwname = nwname + j
        nwnamelist.append(nwname)
    return nwnamelist

ncolnm = space_filler(colnm)

softw = softw.rename(columns= dict(zip(colnm,ncolnm)))
list(softw)
```
